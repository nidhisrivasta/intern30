```python
driver.get("http://www.bcci.tv/")
```

```python
fixture=driver.find_element(By.XPATH,'//a[@class="nav-link.nav-link-fixtures.active"]')
fixture.click()
```

```python
Match_title=[]
Series=[]
Place=[]
Date=[]
Time=[]
```

```python
Match_tags=driver.find_elements(By.XPATH,'//span[@class="matchOrderText ng-binding ng-scope"]')
for i in Match_tags:            
    match=i.text
    Match_title.append(match)
    
title_tags=driver.find_elements(By.XPATH,'//span[@class="ng-binding"]')
for i in title_tags:
    title=i.text
    Series.append(title)
    
place_tags=driver.find_elements(By.XPATH,'//div[@class="fix-place ng-binding ng-scope"]')
for i in place_tags:
    title=i.text
    Place.append(title)  
    
date_tags=driver.find_elements(By.XPATH,'//div[@class="match-card-left match-schedule"]//h5')
for i in date_tags:
    title=i.text
    Date.append(title) 
    
time_tags=driver.find_elements(By.XPATH,'//div[@class="match-card-right match-schedule "]//h5')
for i in time_tags:
    title=i.text
    Time.append(title)    
```

```python
print(len(Match_title),len(Series),len(Place),len(Date),len(Time))
```

```python
df=pd.DataFrame({'Match_title':Match_title,'Series':Series,'Place':Place,'Date':Date,'Time':Time})
df
```

```python
driver.get("http://www.imdb.com/list/ls095964455/")
```

```python
Name=[]
Year_span=[]
Genre=[]
Run_time=[]
Ratings=[]
Votes=[]
```

```python
name_tags=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]//a')
for i in name_tags:            
    nm=i.text
    Name.append(nm)
    
year_tags=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
for i in year_tags:
    yr=i.text
    Year_span.append(yr)
    
genre_tags=driver.find_elements(By.XPATH,'//span[@class="genre"]')
for i in genre_tags:
    title=i.text
    Genre.append(title)  
    
run_tags=driver.find_elements(By.XPATH,'//span[@class="genre"]')
for i in run_tags:
    title=i.text
    Run_time.append(title) 
    
ratings_tags=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]')
for i in ratings_tags:
    title=i.text
    Ratings.append(title)    
    
votes_tags=driver.find_elements(By.XPATH,'//span[@name="nv"]')
for i in votes_tags:
    title=i.text
    Votes.append(title)        
```

```python
print(len(Name),len(Year_span),len(Genre),len(Run_time),len(Ratings),len(Votes))
```

```python
df=pd.DataFrame({'Name':Name,'Year Span':Year_span,'Genre':Genre,'Run Time':Run_time,'Ratings':Ratings,'Votes':Votes})
df
```
