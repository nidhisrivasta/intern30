```python
# Convert all messages to lower case
train_df['comment_text'] = train_df['comment_text'].str.lower()

# Replace email addresses with 'email'
train_df['comment_text'] = train_df['comment_text'].str.replace(r'^.+@[^\.].*\.[a-z]{2,}$',
                                 'emailaddress')

# Replace URLs with 'webaddress'
train_df['comment_text'] = train_df['comment_text'].str.replace(r'^http\://[a-zA-Z0-9\-\.]+\.[a-zA-Z]{2,3}(/\S*)?$',
                                  'webaddress')

# Replace money symbols with 'moneysymb' (Â£ can by typed with ALT key + 156)
train_df['comment_text'] = train_df['comment_text'].str.replace(r'Â£|\$', 'dollers')

# Replace 10 digit phone numbers (formats include paranthesis, spaces, no spaces, dashes) with 'phonenumber'
train_df['comment_text'] = train_df['comment_text'].str.replace(r'^\(?[\d]{3}\)?[\s-]?[\d]{3}[\s-]?[\d]{4}$',
                                  'phonenumber')

    
# Replace numbers with 'numbr'
train_df['comment_text'] = train_df['comment_text'].str.replace(r'\d+(\.\d+)?', 'numbr')


train_df['comment_text'] = train_df['comment_text'].apply(lambda x: ' '.join(
    term for term in x.split() if term not in string.punctuation))

stop_words = set(stopwords.words('english') + ['u', 'Ã¼', 'ur', '4','2', 'im', 'dont', 'doin', 'ure'])
train_df['comment_text'] = train_df['comment_text'].apply(lambda x: ' '.join(
    term for term in x.split() if term not in stop_words))

lem=WordNetLemmatizer()
train_df['comment_text'] = train_df['comment_text'].apply(lambda x: ' '.join(
 lem.lemmatize(t) for t in x.split()))
                                               
                                               
```

```python
y=train_df['bad']
x_train_df,x_test_df,y_train_df,y_test_df=train_test_split(x,y,random_state=56,test_size=.30)
```

```python
y_train_df.shape,y_test_df.shape
```

```python
# LogisticRegression
LG = LogisticRegression(C=1, max_iter = 3000)

LG.fit(x_train_df, y_train_df)

y_pred_train_df = LG.predict(x_train_df)
print('Training accuracy is {}'.format(accuracy_score(y_train_df, y_pred_train_df)))
y_pred_test_df = LG.predict(x_test_df)
print('Test accuracy is {}'.format(accuracy_score(y_test_df,y_pred_test_df)))
print(confusion_matrix(y_test_df,y_pred_test_df))
print(classification_report(y_test_df,y_pred_test_df))
```

```python
# DecisionTreeClassifier
DT = DecisionTreeClassifier()

DT.fit(x_train_df, y_train_df)
y_pred_train_df = DT.predict(x_train_df)
print('Training accuracy is {}'.format(accuracy_score(y_train_df, y_pred_train_df)))
y_pred_test_df = DT.predict(x_test_df)
print('Test accuracy is {}'.format(accuracy_score(y_test_df,y_pred_test_df)))
print(confusion_matrix(y_test_df,y_pred_test_df))
print(classification_report(y_test_df,y_pred_test_df))
```

```python
#RandomForestClassifier
RF = RandomForestClassifier()

RF.fit(x_train_df, y_train_df)
y_pred_train_df = RF.predict(x_train_df)
print('Training accuracy is {}'.format(accuracy_score(y_train_df, y_pred_train_df)))
y_pred_test_df = RF.predict(x_test_df)
print('Test accuracy is {}'.format(accuracy_score(y_test_df,y_pred_test_df)))
print(confusion_matrix(y_test_df,y_pred_test_df))
print(classification_report(y_test_df,y_pred_test_df))
```
